````r
x <- seq(1, 10)
print(x)
````

````r
x <- 1:10
print(x)
````

````r
x <- seq(-pi, pi, length = 50)
print(x)
````

````r
y <- x
f <- outer(x, y, function(x, y) cos(y) / (1 + x^2))
contour(x, y, f)
contour(x, y, f, nlevels = 45, add = T)
````

````r
fa <- (f - t(f)) / 2 # t means the transpose
contour(x, y, fa, nlevels = 15)
````

````r
image(x, y, fa)
````

````r
persp(x, y, fa)
````

````r
persp(x, y, fa, theta = 30)
````

````r
persp(x, y, fa, theta = 30, phi = 20)
````

## Additional Graphical and Numerical Summaries

````r
plot(auto$cylinders, auto$mpg)
````

````r
attach(auto) # to make the variables in this data frame available by name
plot(cylinders, mpg)
````

````r
cylinders <- as.factor(cylinders) # turn it into factors to change plot default
plot(cylinders, mpg, col = "red")
````

````r
plot(cylinders, mpg, col = "red", varwidth = T)
````

````r
plot(cylinders, mpg, col = "red", varwidth = T, horizontal = T)
````

````r
plot(
    cylinders, mpg, 
    col = "red", varwidth = T, xlab = "cylinders", ylab = "MPG",
)
````
