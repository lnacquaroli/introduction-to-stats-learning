(c) What is the mean and standard deviation of each quantitative predictor?

```r
summary(auto)
```

```r
lapply(auto[, -c(ncol(auto))], mean)
```

```r
lapply(auto[, -c(ncol(auto))], sd)
```

```r
auto2 <- auto[-(10:85), ]
```

```r
dim(auto2)
```

```r
lapply(auto2[, -c(ncol(auto2))], mean)
```

```r
lapply(auto2[, -c(ncol(auto2))], sd)
```

It looks like `displacement`, `horsepower`, `year`, have some strong relation with `mpg`. Although `year` has a lot of variablity. 

```r
pairs(~ mpg + displacement + horsepower + year, data = auto)
```
